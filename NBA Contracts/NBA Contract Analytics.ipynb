```python
import pandas as pd
import numpy as np
bold = '\033[1m'
end = '\033[0m'
```

# Importing Data

Load in Latest RAPTOR/WAR values from FiveThirtyEight

```python
player_data = pd.read_csv('https://projects.fivethirtyeight.com/nba-model/2022/latest_RAPTOR_by_team.csv')
player_data.to_csv('Raw Data/latest_raptor.csv')

historical_player_data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-raptor/historical_RAPTOR_by_player.csv')
historical_player_data.to_csv('Raw Data/historical_raptor.csv')
```

```python
relevant_cols = ['player_name', 'player_id', 'season', 'season_type', 'team', 'poss', 'mp', 
                 'raptor_offense', 'raptor_defense', 'raptor_total',
                 'war_total', 'war_reg_season', 'war_playoffs']
player_data = pd.read_csv('Raw Data/latest_raptor.csv')[relevant_cols]
player_data
```

### Import Salary Data
Sourced from basketball-reference.com

```python
# Read in CSV of salaries and limit to relevant information
salary_cap_history = pd.read_csv('Raw Data/salary_cap_history.csv')
salaries = pd.read_csv('Raw Data/salaries_2021.csv')[['Id', 'Tm', '2021-22', 'Signed Using']]
salaries['Signed Using'] = [str(x).lower() for x in salaries['Signed Using']]

# Convert salary figure to integer
salaries = salaries.dropna(subset=['2021-22'])
salaries['2021-22'] = [int(x[1:].strip().replace(',', '')) for x in salaries['2021-22']]

# Merge salary data to the player data via a key of player identifier and team
player_data = player_data.merge(salaries, left_on=['player_id', 'team'], right_on=['Id', 'Tm'])
player_data = player_data.drop(['Id', 'Tm'], axis=1)

player_data
```

### Import and clean NBA executives dataset

Sourced from basketball-reference.com

```python
execs = pd.read_csv('Raw Data/executive_history.csv')
execs.head()
```

Convert executive data to usable format

```python
execs = pd.read_csv('Raw Data/executive_history.csv')
execs.head()

# Explode team years into multiple rows
execs = execs.assign(team_years=execs.team_years.str.split(','))
execs = execs.explode('team_years')

# Clean up and replace the weird characters and split up team / years
team_years = [x.replace('\xa0', ' ').replace('(', '').replace(')', '') \
               .strip().split(' ') for x in execs['team_years']]
execs[['Team', 'Years']] = team_years
execs = execs.drop(['team_years'], axis=1)

# Use the years column to expand the dataframe for each individual year
years = execs[['executive', 'Team', 'Years']].values.tolist()
full_list_execs = []
for i in years:
    
    # If an executive worked for multiple years
    if '-' in i[2]:

        # Define the first and last year of the executive
        first_year, last_year = i[2].split('-')
        if len(last_year) == 2:
            last_year = first_year[:2] + last_year

        # Create an entry for each year in the list
        for j in range(int(first_year), int(last_year)+1):
            full_list_execs.append([i[0], i[1], j])
    else:
        full_list_execs.append(i)

# Convert to dataframe and merge into player data
executives = pd.DataFrame(full_list_execs, columns=['executive', 'team', 'year'])
player_data.merge(executives.loc[executives.year==2022], on='team').drop(['year'], axis=1)
executives.head(3)
```

# Analysis

Lets start with a simple measure to see if salary is predictive of WAR and RAPTOR

```python
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

def show_correlation_and_regress(df_in, factor, RS_PO):
    
    # Segment data to regular/post season only
    df_seg = df_in.loc[df_in.season_type==RS_PO]
    
    
    # Correlation measures
    print(bold+'Below are two common correlation measures, Pearson and Spearman.'+end)
    print('Pearson is best for linear correlations, while Spearman fits to non-linear data.\n')
    
    corr_p, _ = stats.pearsonr(df_seg['2021-22'], df_seg[factor])
    print('Pearsons Correlation:', round(corr_p,2))
    
    corr_s, _ = stats.spearmanr(df_seg['2021-22'], df_seg[factor])
    print('Spearman Correlation:', round(corr_s,2))
    
    # Plot regression plot
    sns.regplot(x=df_seg['2021-22'], y=df_seg[factor])
    plt.show()
    
    # Remove outliers
    print(bold+'There appear to be three really bad players, all with extremely low salaries.'+end)
    print('Since these contracts are not indicative of ability, we can remove them as outliers.')
    print('The Pearson correlation improves heavily here while the Spearman does not.')
    print('This is because Spearman is reliant on rank correlation, while Pearson is purely linear.')
    print('We get a much better picture of what is actually going on here now.\n')
    
    # Correlation measures without outliers
    df_seg = df_seg.loc[df_seg[factor]>-20]
    
    corr_p, _ = stats.pearsonr(df_seg['2021-22'], df_seg[factor])
    print('Pearsons Correlation:', round(corr_p,2))
    
    corr_s, _ = stats.spearmanr(df_seg['2021-22'], df_seg[factor])
    print('Spearman Correlation:', round(corr_s,2))
    
    # Plot regression plot
    sns.regplot(x=df_seg['2021-22'], y=df_seg[factor])
    plt.show()
    
    print(bold+'So, what do we get from this?'+end)
    print('Simple, that there is a correlation between salary and ability.')
    print('Good management should be able to do better than average.')
    
show_correlation_and_regress(player_data, 'raptor_total', 'RS')
```
